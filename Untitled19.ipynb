```python
import pandas as pd
```

```python
df = pd.read_csv('/content/ratings.csv')
```

```python
df
```

Переведём данные из столбца timestamp в формат datetime

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], unit = 's')
```

Создадим DataFrame active_users, в который мы занесём резульат группировки по столбцу userId, который показывает количество оценок каждого пользователя за всю жизнь

```python
active_users = df.groupby('userId').size().reset_index(name = 'count')
```

```python
active_users
```

```python
df_count = df.merge(active_users, on='userId', how='left')
```

Отфильтровываем данные и снова группируем их, для того чтобы получить DataFrame под названием live_time. В нём будет только userId, самая первая дата и самя последняя дата

```python
live_time = df_count.loc[df_count['count'] > 100].groupby('userId')['timestamp'].agg(['min', 'max'])
```

Надём разницу между последней и первой датой. Занесём её в столбец live. С помощью .dt.days, преобразуем результат в количетсво дней

```python
live_time['live'] = (live_time['max'] - live_time['min']).dt.days
```

Далее просто находим среднее

```python
average_time = live_time['live'].mean()
```
